```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python

# Reading each CSV file into separate DataFrames with mnemonic names
inflation_data = pd.read_csv(
    "C:/Users/Sayan/Downloads/archive (3)/9 - India_Inflation_Rate_Historical_Data.csv", encoding='ISO-8859-1'
).drop(columns=['Unnamed: 0'])

economic_growth_data = pd.read_csv(
    "C:/Users/Sayan/Downloads/archive (3)/1 - India_Economic_Growth_Historical_Data.csv", encoding='ISO-8859-1'
).drop(columns=['Unnamed: 0'])

gdp_growth_data = pd.read_csv(
    "C:/Users/Sayan/Downloads/archive (3)/2 -India_GDP_Growth_Rate_Historical_Data.csv", encoding='ISO-8859-1'
).drop(columns=['Unnamed: 0'])

gdp_per_capita_data = pd.read_csv(
    "C:/Users/Sayan/Downloads/archive (3)/4 - India_GDP_Per_Capita_Historical_Data.csv", encoding='ISO-8859-1'
).drop(columns=['Unnamed: 0'])

gni_data = pd.read_csv(
    "C:/Users/Sayan/Downloads/archive (3)/5 - India_GNI_Historical_Data.csv", encoding='ISO-8859-1'
).drop(columns=['Unnamed: 0'])

gni_per_capita_data = pd.read_csv(
    "C:/Users/Sayan/Downloads/archive (3)/6 - India_GNI_Per_Capita_Historical_Data.csv", encoding='ISO-8859-1'
).drop(columns=['Unnamed: 0'])

gnp_data = pd.read_csv(
    "C:/Users/Sayan/Downloads/archive (3)/7 - India_GNP_Historical_Data.csv", encoding='ISO-8859-1'
).drop(columns=['Unnamed: 0'])

gov_debt_data = pd.read_csv(
    "C:/Users/Sayan/Downloads/archive (3)/8 - India_Government_Debt_as_percent_of_GDP_Historical_Data.csv", encoding='ISO-8859-1'
).drop(columns=['Unnamed: 0'])

manufacturing_output_data = pd.read_csv(
    "C:/Users/Sayan/Downloads/archive (3)/10 - India_Manufacturing_Output_Historical_Data.csv", encoding='ISO-8859-1'
).drop(columns=['Unnamed: 0'])
```

```python
print("\nGovernment Debt Data:")
print(gov_debt_data.head())
gov_debt_data['Government_Debt_as_percent_of_GDP'] = gov_debt_data['Government_Debt_as_percent_of_GDP'].str.replace('%', '').astype(float)
gov_debt_data['Annual_percent_geowth'] = gov_debt_data['Annual_percent_geowth'].str.replace('%', '').astype(float)
gov_debt_data.head()
# Step 4: Plot the data
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting Government_Debt_as_percent_of_GDP
ax1.plot(gov_debt_data['year'], gov_debt_data['Government_Debt_as_percent_of_GDP'], color='tab:red', marker='o', label='Government Debt as % of GDP')
ax1.set_xlabel('Year')
ax1.set_ylabel('Government Debt as % of GDP', color='tab:red')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Creating a second y-axis for Annual_percent_geowth
ax2 = ax1.twinx()
ax2.plot(gov_debt_data['year'], gov_debt_data['Annual_percent_geowth'], color='tab:blue', marker='x', linestyle='--', label='Annual Percent Growth')
ax2.set_ylabel('Annual Percent Growth (%)', color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Adding title and grid
plt.suptitle('Government Debt Data Over Time')  # Use plt.suptitle() instead of plt.title()
fig.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout to make room for the title
plt.grid(True)

# Adding a legend
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

# Show the plot
plt.show()
```

```python
print("\nManufacturing Output Data:")
print(manufacturing_output_data.head())
manufacturing_output_data['billions_of_us'] = manufacturing_output_data['billions_of_us'].replace({'\$': '', 'B': ''}, regex=True).astype(float)
manufacturing_output_data['percent_of_GDP'] = manufacturing_output_data['percent_of_GDP'].str.replace('%', '').astype(float)
manufacturing_output_data.head()
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting billions_of_us
ax1.plot(manufacturing_output_data['year'], manufacturing_output_data['billions_of_us'], color='tab:blue', marker='o', label='Manufacturing Output (Billions of $US)')
ax1.set_xlabel('Year')
ax1.set_ylabel('Billions of $US', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Creating a second y-axis for percent_of_GDP
ax2 = ax1.twinx()
ax2.plot(manufacturing_output_data['year'], manufacturing_output_data['percent_of_GDP'], color='tab:green', marker='x', linestyle='--', label='Percent of GDP')
ax2.set_ylabel('Percent of GDP (%)', color='tab:green')
ax2.tick_params(axis='y', labelcolor='tab:green')

# Adding title and grid
plt.suptitle("Manufacturing Output Data Over Time")
fig.tight_layout()
#plt.grid(True)

# Adding a legend
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

# Show the plot
plt.show()
```

```python
print("Inflation Data:")
print(inflation_data.head())
inflation_data['Inflation_Rate'] = inflation_data['Inflation_Rate'].str.replace('%', '').astype(float)
inflation_data['Annual_percent_geowth'] = inflation_data['Annual_percent_geowth'].str.replace('%', '').astype(float)

# Step 4: Plot the data
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting Inflation_Rate
ax1.plot(inflation_data['year'], inflation_data['Inflation_Rate'], color='tab:red', marker='o', label='Government Debt as % of GDP')
ax1.set_xlabel('Year')
ax1.set_ylabel('Government Debt as % of GDP', color='tab:red')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Creating a second y-axis for Annual_percent_geowth
ax2 = ax1.twinx()
ax2.plot(inflation_data['year'], inflation_data['Annual_percent_geowth'], color='tab:blue', marker='x', linestyle='--', label='Annual Percent Growth')
ax2.set_ylabel('Annual Percent Growth (%)', color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Adding title and grid
plt.suptitle('Government Debt Data Over Time')  # Use plt.suptitle() instead of plt.title()
fig.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout to make room for the title
plt.grid(True)

# Adding a legend
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

# Show the plot
plt.show()
```

```python
print("\nEconomic Growth Data:")
print(economic_growth_data.head())
economic_growth_data['GDP'] = economic_growth_data['GDP'].replace({'\$': '', 'B': '', ',': ''}, regex=True).astype(float)
economic_growth_data['Per_Capita'] = economic_growth_data['Per_Capita'].replace({'\$': '', ',': ''}, regex=True).astype(float)
economic_growth_data['Annual_percent_geowth'] = economic_growth_data['Annual_percent_geowth'].str.replace('%', '').astype(float)

# Step 4: Plot the data
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting GDP
ax1.plot(economic_growth_data['year'], economic_growth_data['GDP'], color='tab:red', marker='o', label='GDP (Billions of $US)')
ax1.set_xlabel('Year')
ax1.set_ylabel('GDP (Billions of $US)', color='tab:red')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Creating a second y-axis for Per_Capita
ax2 = ax1.twinx()
ax2.plot(economic_growth_data['year'], economic_growth_data['Per_Capita'], color='tab:blue', marker='x', linestyle='--', label='Per Capita ($US)')
ax2.set_ylabel('Per Capita ($US)', color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Creating a third y-axis for Annual_percent_geowth
ax3 = ax1.twinx()
ax3.spines["right"].set_position(("outward", 60))
ax3.plot(economic_growth_data['year'], economic_growth_data['Annual_percent_geowth'], color='tab:green', marker='s', linestyle='-.', label='Annual Percent Growth')
ax3.set_ylabel('Annual Percent Growth (%)', color='tab:green')
ax3.tick_params(axis='y', labelcolor='tab:green')

# Adding title and grid
plt.suptitle('Economic Growth Data Over Time')  # Use plt.suptitle() instead of plt.title()
fig.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout to make room for the title
plt.grid(True)

# Adding a legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines3, labels3 = ax3.get_legend_handles_labels()
ax3.legend(lines1 + lines2 + lines3, labels1 + labels2 + labels3, loc='upper left')

# Show the plot
plt.show()
```
